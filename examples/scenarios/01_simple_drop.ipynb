# Example 01: Simple Drop Test

This notebook demonstrates how to use the high-level `Scenario` API to define a simple falling object test with minimal code.

## Setup
We start by configuring the environment to display plots inline.

```python
%matplotlib inline
import sys
from pathlib import Path

# Setup path for local development (if aerislab is not installed in site-packages)
sys.path.insert(0, str(Path.cwd().parent.parent / "src"))

from aerislab.api.scenario import Scenario
from aerislab.components.standard import Payload
```

## 1. Define the Payload

We use the `Payload` standard component which handles rigid body creation, inertia, and basic aerodynamics automatically.

```python
capsule = Payload(
    name="capsule_NB", 
    mass=10.0,      # kg
    radius=0.5,     # m
    Cd=0.5,         # Drag coefficient
    position=[0, 0, 1000], # Start at 1km altitude
    velocity=[0, 0, 0]     # Start from rest
)
```

## 2. Run Simulation

The `Scenario` manager orchestrates the physics world and solver. By setting `show=True` in `enable_plotting`, graphs will be displayed inline in the notebook.

```python
scenario = Scenario(name="01_simple_drop_NB") \
    .add_system([capsule]) \
    .enable_plotting(show=True) \
    .run(duration=30.0)

print(f"Results saved to: {scenario.world.output_path}")
```
